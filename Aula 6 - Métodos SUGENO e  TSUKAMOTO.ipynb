```python
valores = np.arange(-5,5,0.1)
X1 = fuzz.sigmf(valores,0,-2.5) #função Sigmoide
X2 = fuzz.sigmf(valores,0,2.5)
fig,ax = plt.subplots()
ax.plot(valores,X1,'m',linewidth=3,label= 'Small')
ax.plot(valores,X2,'y',linewidth=3,label='Large') 
ax.set_ylabel('grau de pertinência')
ax.set_xlabel('Valores da Variável de Entrada')
ax.set_ylim(-0.05, 1.05);
ax.grid(True)
ax.legend(loc='upper right')
ax.legend(bbox_to_anchor=(1.60, 1.15))
plt.show()

```

$$ \Large{\text{Método TSUKAMOTO de Defuzzificação}}$$

<img src="Images/TSUKAMOTO.png">

```python
intervaloX = np.arange(-10,10.1,0.1)
intervaloY = np.arange(0,10.1,0.1)
Small = fuzz.gaussmf(intervaloX,-10,5.5) #função Sigmoide
Medium = fuzz.gaussmf(intervaloX,0,3.5)
Large = fuzz.gaussmf(intervaloX,10,5.5)
fig,ax = plt.subplots()
ax.plot(intervaloX,Small,'r',linewidth=3,label= 'Small')
ax.plot(intervaloX,Medium,'g',linewidth=3,label='Medium')
ax.plot(intervaloX,Large,'b',linewidth=3,label='Large')

ax.set_ylabel('grau de pertinência')
ax.set_xlabel('Valores da Variável de Entrada')
ax.set_ylim(-0.05, 1.05);
ax.grid(True)
ax.legend(loc='upper right')
ax.legend(bbox_to_anchor=(1.60, 1.15))
plt.show()

```

```python
C1 = fuzz.trapmf(intervaloY,[0,2,10,10])
C2 = fuzz.trapmf(intervaloY,[0,0,4.5,6])
C3 = fuzz.trapmf(intervaloY,[7.4,10,10,10])
fig,ax = plt.subplots()
ax.plot(intervaloY,C1,'b',linewidth=3,label= 'Small')
ax.plot(intervaloY,C2,'g--',linewidth=3,label='Medium')
ax.plot(intervaloY,C3,'m*',linewidth=3,label='Large')

ax.set_ylabel('grau de pertinência')
ax.set_xlabel('Valores da Variável de Saída')
ax.set_ylim(-0.05, 1.05);
ax.grid(True)
ax.legend(loc='upper right')
ax.legend(bbox_to_anchor=(1.60, 1.15))
plt.show()
```

$$\Large{\text{ IF X is small , THEN z is C1}\\
 \text{IF X is medium,THEM z is C2 }\\
\text{IF X is large, THEN z is C3}\\}$$

```python
#função para achar a posição
def posicao(array,value):
    idx = (np.abs(array-value)).argmin()
    return array[idx]
```

```python
w1 = Small[47]
w2 = Medium[47]
w3 = Large[47]
p1 = posicao(C1,w1)
p2 = posicao(C2,w2)
p3 = posicao(C3,w3)
P1 = np.where(C1 ==p1 )   
D1 = P1[0][0]
P2 = np.where(C2 ==p2 )   
D2 = P2[0][0]
P3 = np.where(C3 ==p3 )   
D3 = P3[0][0]
```

```python
Zfinal = (( w1* D1/10 ) + (w2*D2/10) + (w3*D3/10)) / (w1+w2+w3)
Zfinal
```
