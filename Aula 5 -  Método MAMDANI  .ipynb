```python
#importar os pacotes necessários
import numpy as np
import skfuzzy as fuzz
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
from matplotlib.font_manager import FontProperties
%matplotlib inline 
```

# Método Mamdani MaxMin

$$ \large{\text{ Exemplo 5.15 Adaptado }} $$

```python
#considere os seguintes intervalos operacionais para as variáveis fuzzy ( ANTECEDENTES)
intervalo1 =  np.arange(1,11,0.1)
intervalo2 = np.arange(1,11,0.1)
#intevalor para o consequente
intervaloB = np.arange(1,11,0.1)
```

```python
#definir as variáveis fuzzy

A11 = fuzz.trimf(intervalo1,[2,5,7]) #baixa massa
A21 = fuzz.trimf(intervalo1,[4,7,10]) #média massa 

A12 = fuzz.trimf(intervalo2,[1,4,7]) #velocidade alta
A22 = fuzz.trimf(intervalo2,[3,6,9]) #velocidade média

B1 = fuzz.trimf(intervaloB,[1,5,8])
B2 = fuzz.trimf(intervaloB,[4,7,9])
```

```python
#MEDIDA DE MASSA
fig,ax = plt.subplots()
ax.plot(intervalo1,A11,'m--',linewidth=3,label='Variável Fuzzy A11')
ax.plot(intervalo1,A21,'k--',linewidth=3,label='Variável Fuzzy A21')

ax.set_ylabel('grau de pertinência')
ax.set_xlabel('Valores')
ax.set_ylim(-0.05, 1.05);

ax.legend(loc='upper right')
ax.legend(bbox_to_anchor=(1.80, 1.15))
ax.grid(True)
plt.xlim(1,12,1.0)
plt.xticks(range(1,12))
plt.show()
```

```python
#MEDIDA DE VELOCIDADE
fig,ax = plt.subplots()
ax.plot(intervalo2,A12,'b--',linewidth=3,label='Variável Fuzzy A12')
ax.plot(intervalo2,A22,'g--',linewidth=3,label='Variável Fuzzy A22')

ax.set_ylabel('grau de pertinência')
ax.set_xlabel('Valores')
ax.set_ylim(-0.05, 1.05);

ax.legend(loc='upper right')
ax.legend(bbox_to_anchor=(1.80, 1.15))
ax.grid(True)
plt.xlim(1,12,1.0)
plt.xticks(range(1,12))
plt.show()
```

```python
# MEDIDA DE ENERGIA
fig,ax = plt.subplots()
ax.plot(intervaloB,B1,'k--',linewidth=3,label='Variável Fuzzy B1')
ax.plot(intervaloB,B2,'y--',linewidth=3,label='Variável Fuzzy B2')

ax.set_ylabel('grau de pertinência')
ax.set_xlabel('Valores')
ax.set_ylim(-0.05, 1.05);

ax.legend(loc='upper right')
ax.legend(bbox_to_anchor=(1.80, 1.15))
ax.grid(True)
plt.xlim(1,12,1.0)
plt.xticks(range(1,12))
plt.show()
```

$$ \Large{ \text{ Agora vamos agregar usando o oparador de Máximo } }$$

```python
NovoB =  fuzz.fuzzy_or(intervaloB,B1,intervaloB,B2)
```

```python
NovoB
```

```python
fig,ax = plt.subplots()
ax.plot(intervaloB,B1,'k--',linewidth=3,label=' Nova Variável Fuzzy B1')
ax.plot(intervaloB,B2,'y--',linewidth=3,label=' Nova Variável Fuzzy B2')
ax.plot(NovoB[0],NovoB[1],'b--',linewidth=3,label=' Nova Variavel B MAMDANI')
ax.fill(NovoB[0],NovoB[1],"m")


ax.set_ylabel('grau de pertinencia')
ax.set_xlabel('Valores')
ax.set_ylim(-0.05, 1.05);

ax.legend(loc='upper right')
ax.legend(bbox_to_anchor=(1.80, 1.15))
ax.grid(True)
plt.xlim(1,10,1.0)
plt.xticks(range(1,10))
plt.show()
```

### Logo, o valor de y* de acordo com as regras apresentadas e com os inputs X e Y será: 

```python
defuzz_centroid = fuzz.defuzz(NovoB[0], NovoB[1], 'centroid')  # Same as skfuzzy.centroid
#defuzz_bisector = fuzz.defuzz(R49[0], R49[1], 'bisector')
#defuzz_mom = fuzz.defuzz(R49[0], R49[1], 'mom')
#defuzz_som = fuzz.defuzz(R49[0], R49[1], 'som')
#defuzz_lom = fuzz.defuzz(R49[0], R49[1], 'lom')
#
# Collect info for vertical lines
labels = ['centroide'] #'bissetriz', 'média dos máximos', 'mínimo dos máximos',
          #'máximo dos máximos' 
xvals = [defuzz_centroid]
         #defuzz_bisector,
         #defuzz_mom,
         #defuzz_som,
         #defuzz_lom]
colors = ['r', 'b', 'g', 'c', 'm']
ymax = [fuzz.interp_membership(NovoB[0], NovoB[1], i) for i in xvals]

# Display and compare defuzzification results against membership function
plt.figure(figsize=(8, 5))

plt.plot(NovoB[0], NovoB[1], 'k')
for xv, y, label, color in zip(xvals, ymax, labels, colors):
    plt.vlines(xv, 0, y, label=label, color=color)
plt.ylabel('Fuzzy membership')
plt.xlabel('Universe variable (arb)')
ax.set_xlim(0.0,101.05)

plt.ylim(-0.01, 1.1)
plt.legend(loc=2)

plt.show()
```
