```python
#importar os pacotes necessários
import numpy as np
import skfuzzy as fuzz
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
from matplotlib.font_manager import FontProperties
%matplotlib inline 
```

```python
X = np.arange(0.0, 6.0, 1) #intervalo da variável temperatura
valores=[]
valores = list(range(6))
A = []
B = []
C = []
```

```python
for i in valores:
    k=abs(X[i])
    A.append(1/(1+(5*((k-5)**2))))
    
for i in valores:
    k=abs(X[i])
    B.append(2**(-k))
    
for i in valores:
    k=abs(X[i])
    C.append(2*k/(k+5))
```

```python
A = np.array(A)
B = np.array(B)
C = np.array(C)
```

```python
fig,ax = plt.subplots()
ax.plot(X,A,'b--',linewidth=3,label='Função A')
ax.plot(X,B,'r--',linewidth=3,label='Função B')
ax.plot(X,C,'k--',linewidth=3,label='Função C')

ax.set_ylabel('grau de pertinência')
ax.set_xlabel('Intervalo da Variável X')
ax.set_ylim(-0.05, 1.05);

ax.legend(loc='upper right')
ax.legend(bbox_to_anchor=(1.60, 1.15))
plt.show()

```

```python
#Realizar o Lambda Cut para 0.2 - 0.6 - 0.9 - 1.0 

Cut1A = fuzz.arglcut(A, 0.2)
Cut1B = fuzz.arglcut(B, 0.2)
Cut1C = fuzz.arglcut(C, 0.2)

Cut2A = fuzz.arglcut(A, 0.6)
Cut2B = fuzz.arglcut(B, 0.6)
Cut2C = fuzz.arglcut(C, 0.6)

Cut3A = fuzz.arglcut(A, 0.9)
Cut3B = fuzz.arglcut(B, 0.9)
Cut3C = fuzz.arglcut(C, 0.9)

Cut4A = fuzz.arglcut(A, 1.0)
Cut4B = fuzz.arglcut(B, 1.0)
Cut4C = fuzz.arglcut(C, 1.0)
```

```python
Z = np.arange(0.0, 7, 1) #interlavo da variável Z (quilos)
A1 = fuzz.trimf(Z,[0,3,4])
A2 = fuzz.trapmf(Z,[2,3,5,6])
```

```python
fig,ax = plt.subplots()
ax.plot(Z,A1,'b--',linewidth=3,label='Mistura A1')


ax.set_ylabel('grau de pertinência')
ax.set_xlabel('Intervalo da Variável Z')
ax.set_ylim(-0.05, 1.05);
ax.set_xlim(-0.05,6.05)
ax.legend(loc='upper right')
ax.legend(bbox_to_anchor=(1.60, 1.15))
plt.show()





fig,ax = plt.subplots()
ax.plot(Z,A2,'r--',linewidth=3,label='Mistura A2')


ax.set_ylabel('grau de pertinência')
ax.set_xlabel('Intervalo da Variável Z')
ax.set_ylim(-0.05, 1.05);
ax.set_xlim(-0.05,6.25);
ax.legend(loc='upper right')
ax.legend(bbox_to_anchor=(1.60, 1.15))
plt.show()
```

```python
R49 = fuzz.fuzzy_or(Z,A1,Z,A2) 
```

```python
fig,ax = plt.subplots()
ax.plot(R49[0],R49[1],'k',linewidth=5,label='Mistura A1 e A2')


ax.set_ylabel('grau de pertinência')
ax.set_xlabel('Intervalo da Variável Z')
ax.set_ylim(-0.05, 1.05);
ax.set_xlim(-0.05,6.25);
ax.legend(loc='upper right')
ax.legend(bbox_to_anchor=(1.60, 1.15))
plt.show()
```

```python
#calculando a deffuzificação pela média ponderada 
k = fuzz.trimf(Z,[0,1,2])
for i in range(7):
    k[i] =  ( R49[0][i]*R49[1][i])
```

```python
fig,ax = plt.subplots()
ax.plot(R49[0],R49[1],'k',linewidth=5,label='Mistura A1 e A2')
plt.axvline(x=3.0555)



ax.set_ylabel('grau de pertinência')
ax.set_xlabel('Intervalo da Variável Z')
ax.set_ylim(-0.05, 1.05);
ax.set_xlim(-0.05,6.25);
ax.legend(loc='upper right')
ax.legend(bbox_to_anchor=(1.60, 1.15))
plt.show()
```

```python
defuzz_centroid = fuzz.defuzz(R49[0], R49[1], 'centroid')  # Same as skfuzzy.centroid
defuzz_bisector = fuzz.defuzz(R49[0], R49[1], 'bisector')
defuzz_mom = fuzz.defuzz(R49[0], R49[1], 'mom')
defuzz_som = fuzz.defuzz(R49[0], R49[1], 'som')
defuzz_lom = fuzz.defuzz(R49[0], R49[1], 'lom')

# Collect info for vertical lines
labels = ['centroide', 'bissetriz', 'média dos máximos', 'mínimo dos máximos',
          'máximo dos maáximos']
xvals = [defuzz_centroid,
         defuzz_bisector,
         defuzz_mom,
         defuzz_som,
         defuzz_lom]
colors = ['r', 'b', 'g', 'c', 'm']
ymax = [fuzz.interp_membership(R49[0], R49[1], i) for i in xvals]

# Display and compare defuzzification results against membership function
plt.figure(figsize=(8, 5))

plt.plot(R49[0], R49[1], 'k')
for xv, y, label, color in zip(xvals, ymax, labels, colors):
    plt.vlines(xv, 0, y, label=label, color=color)
    
#plt.axvline(x=3.0555)
plt.ylabel('Fuzzy membership')
plt.xlabel('Universe variable (arb)')
plt.ylim(-0.01, 1.1)
plt.legend(loc=2)

plt.show()

```
